```python
from ddgs import DDGS
import requests, os, shutil

folder = "kebab_images"
shutil.rmtree(folder, ignore_errors=True)
os.makedirs(folder, exist_ok=True)

TARGET = 100
saved = 0

with DDGS() as ddgs:
    results = ddgs.images("kebab shashlik food", max_results=300)

    for r in results:
        if saved >= TARGET:
            break
        try:
            img = requests.get(r["image"], timeout=10)
            if img.status_code == 200:
                with open(f"{folder}/{saved}.jpg", "wb") as f:
                    f.write(img.content)
                saved += 1
                print(saved, "ok")
        except:
            pass

print("Jami yuklandi:", saved)
```
